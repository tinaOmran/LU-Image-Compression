```python
import numpy as np
import cv2
from scipy.linalg import lu, svd
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim
import matplotlib.pyplot as plt
import os
from scipy.sparse import csr_matrix, issparse
from scipy.sparse import csr_matrix
from scipy.linalg import lu
```

```python
# --- LU Decomposition ---
def lu_decomposition_no_pivoting(A):
    A = A.copy().astype(np.float64)
    n = A.shape[0]
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        for j in range(i, n):
            U[i, j] = A[i, j] - np.dot(L[i, :i], U[:i, j])

        L[i, i] = 1
        for j in range(i+1, n):
            if U[i, i] == 0:
                raise ValueError("Division by zero: LU without pivoting failed.")
            L[j, i] = (A[j, i] - np.dot(L[j, :i], U[:i, i])) / U[i, i]

    return L, U
```

```python
def lu_decomposition(A):
    P1, L1, U1 = lu(A)
    return P1, L1, U1

def lu_decomposition_with_pivot(A):
    A = A.copy().astype(float)
    n = A.shape[0]
    
    P = np.eye(n)
    L = np.zeros((n, n))
    U = A.copy()

    for k in range(n):
        # پیدا کردن اندیس ماکزیمم مقدار در ستون k از سطر k به پایین
        max_row = np.argmax(np.abs(U[k:, k])) + k
        
        # تعویض سطرها در U و P و L (بخش‌های قبلی)
        if max_row != k:
            U[[k, max_row], :] = U[[max_row, k], :]
            P[[k, max_row], :] = P[[max_row, k], :]
            if k > 0:
                L[[k, max_row], :k] = L[[max_row, k], :k]
        
        # ساخت L و U
        for i in range(k+1, n):
            if U[k, k] == 0:
                raise ValueError("Zero pivot encountered.")
            L[i, k] = U[i, k] / U[k, k]
            U[i, :] -= L[i, k] * U[k, :]
    
    np.fill_diagonal(L, 1)
    return P, L, U

```

```python
# --- Quantization ---
def quantize_and_filter(image, levels=16, block_size=8, threshold=10):
    max_val = 255
    bins = np.linspace(0, max_val, levels + 1)
    quantized = np.digitize(image, bins) - 1
    centers = (bins[:-1] + bins[1:]) / 2
    compressed = centers[quantized.astype(int)]

    filtered = np.copy(compressed)
    h, w = image.shape
    for i in range(0, h, block_size):
        for j in range(0, w, block_size):
            block = image[i:i+block_size, j:j+block_size]
            if block.size == 0:
                continue
            std = np.std(block)
            if std < threshold:
                filtered[i:i+block_size, j:j+block_size] = 0

    return filtered
```

```python
# --- LU+Threshold Encoding ---
def lu_threshold(matrix, threshold=1e-3):
    matrix[np.abs(matrix) < threshold] = 0
    return matrix
```

```python
# --- Encoder ---
def encoder(image, method="LU", **kwargs):
    if method == "LU":
        k = kwargs.get("k", 10)
        image_mod = image.copy()
        image_mod += np.eye(image_mod.shape[0]) * 1e-5
        L, U = lu_decomposition_no_pivoting(image_mod)
        L_k = csr_matrix(L[:, :k])
        U_k = csr_matrix(U[:k, :])
        compressed = (L_k, U_k)

    elif method == "LU+Threshold":
        k = kwargs.get("k", 10)
        threshold = kwargs.get("threshold", 1e-3)
        image_mod = image.copy()
        image_mod += np.eye(image_mod.shape[0]) * 1e-5
        L, U = lu_decomposition_no_pivoting(image_mod)
        L = lu_threshold(L, threshold)
        U = lu_threshold(U, threshold)
        L_k = csr_matrix(L[:, :k])
        U_k = csr_matrix(U[:k, :])
        compressed = (L_k, U_k)

    elif method == "LU with pivot":
        k = kwargs.get("k", 10)
        image_mod = image.copy()
        image_mod += np.eye(image_mod.shape[0]) * 1e-5
        P, L, U = lu_decomposition(image_mod)
        P_k = csr_matrix(P)         # ← تبدیل به sparse
        L_k = csr_matrix(L[:, :k])
        U_k = csr_matrix(U[:k, :])
        compressed = (P_k, L_k, U_k)

    elif method == "SVD":
        k = kwargs.get("k", 10)
        U, S, Vt = svd(image, full_matrices=False)
        S_k = np.diag(S[:k])
        compressed = (U[:, :k], S_k, Vt[:k, :])

    elif method == "JPEG":
        cv2.imwrite("compressed.jpg", image.astype(np.uint8), [int(cv2.IMWRITE_JPEG_QUALITY), kwargs.get("quality", 30)])
        compressed = "compressed.jpg"

    elif method == "Quantized":
        levels = kwargs.get("levels", 16)
        block_size = kwargs.get("block_size", 5)
        threshold = kwargs.get("threshold", 0.05)
        compressed = quantize_and_filter(image, levels, block_size, threshold)

    elif method == "LU+Quantized":
        k = kwargs.get("k", 10)
        levels = kwargs.get("levels", 16)
        block_size = kwargs.get("block_size", 5)
        threshold = kwargs.get("threshold", 0.05)
        # ابتدا کوانتیزه و فیلتر کن
        quantized_img = quantize_and_filter(image, levels, block_size, threshold)
        quantized_img += np.eye(quantized_img.shape[0]) * 1e-5
        # بعد تجزیه LU بدون pivot انجام بده
        L, U = lu_decomposition_no_pivoting(quantized_img)
        L_k = L[:, :k]
        U_k = U[:k, :]
        # تبدیل ماتریس‌ها به اسپارس
        L_sparse = csr_matrix(L_k)
        U_sparse = csr_matrix(U_k)
        compressed = (L_sparse, U_sparse)


    return compressed
```

```python
# --- Decoder ---
from scipy.sparse import issparse

def decoder(compressed, method="LU"):
    if method in ["LU", "LU+Threshold", "LU+Quantized"]:
        L_k, U_k = compressed
        return (L_k @ U_k).toarray() if issparse(L_k) and issparse(U_k) else L_k @ U_k
    elif method == "LU with pivot":
        P_k, L_k, U_k = compressed
        result = (L_k @ U_k)
        if issparse(result):
            result = result.toarray()
        return P_k @ result if not issparse(P_k) else P_k.dot(result)
    elif method == "SVD":
        U_k, S_k, Vt_k = compressed
        return U_k @ S_k @ Vt_k
    elif method == "JPEG":
        return cv2.imread(compressed, cv2.IMREAD_GRAYSCALE).astype(np.float64)
    elif method == "Quantized":
        return compressed
    
```

```python
# --- Compression Ratio ---
def compute_compression_ratio(original, compressed, method, **kwargs):
    original_size = original.size * original.itemsize

    if method == "LU":
        L_k, U_k = compressed
        compressed_size = (L_k.size + U_k.size) * original.itemsize

    elif method == "LU with pivot":
        P_k, L_k, U_k = compressed
        compressed_size = (P_k.size + L_k.size + U_k.size) * original.itemsize

    elif method == "SVD":
        U_k, S_k, Vt_k = compressed
        compressed_size = (U_k.size + S_k.size + Vt_k.size) * original.itemsize

    elif method == "JPEG":
        filename = compressed
        compressed_size = os.path.getsize(filename)

    elif method == "Quantized":
        levels = kwargs.get("levels", 16)
        bits_per_pixel = int(np.ceil(np.log2(levels)))
        compressed_size = original.size * bits_per_pixel / 8

    elif method == "LU+Quantized":
        L_k, U_k = compressed
        compressed_size = (L_k.size + U_k.size) * original.itemsize

    elif method == "LU+Threshold":
        L_k, U_k = compressed
        compressed_size = (L_k.size + U_k.size) * original.itemsize

    ratio = compressed_size / original_size
    print(f"{method} -> Compression Ratio: {ratio:.2f}\n")
    return ratio
```

```python
# --- Quality Evaluation ---
def evaluate_quality(original, reconstructed, method_name):
    reconstructed = np.clip(reconstructed, 0, 255)
    psnr_val = psnr(original, reconstructed, data_range=255)
    ssim_val = ssim(original, reconstructed, data_range=255)
    print(f"{method_name} -> PSNR: {psnr_val:.2f} dB,\t SSIM: {ssim_val:.4f}\n")
    return reconstructed
```

```python
# ---Load first Image ---
image = cv2.imread("images (1).jpg", cv2.IMREAD_GRAYSCALE)
image = image.astype(np.float64)
image = cv2.resize(image, (512, 512))
```

```python
# --- Run Experiments ---
methods = [
    ("LU", {"k": 512}),
    ("SVD", {"k": 10}),
    ("LU with pivot", {"k": 512}),
    ("JPEG", {"quality": 30}),
    ("Quantized", {"levels": 16}),
    ("LU+Quantized", {"k": 512, "levels": 32, "block_size": 5, "threshold": 0.05}),
    ("LU+Threshold", {"k": 512, "threshold": 5e-5})  # ← روش جدید
]

titles = ["Original"]
images = [np.clip(image, 0, 255)]

for method, params in methods:
    comp = encoder(image, method, **params)
    recon = decoder(comp, method)
    recon_clip = evaluate_quality(image, recon, method)
    compute_compression_ratio(image, comp, method, **params)

    images.append(recon_clip)
    if method == "SVD":
        titles.append(f"SVD (k={params['k']})")
    elif method == "LU":
        titles.append(f"LU (k={params['k']})")
    elif method == "LU with pivot":
        titles.append(f"LU with pivot (k={params['k']})")
    elif method == "JPEG":
        titles.append(f"JPEG (Q={params['quality']})")
    elif method == "Quantized":
        titles.append(f"Quantized ({params['levels']} levels)")
    elif method == "LU+Quantized":
        titles.append(f"LU+Quantized (k={params['k']}, levels={params['levels']})")
    elif method == "LU+Threshold":
        titles.append(f"LU+Threshold (k={params['k']})")
```

```python
# --- Display Results ---
plt.figure(figsize=(18, 10))
#plt.figure(figsize=(4 * n_cols, 4 * n_rows)
for i in range(len(images)):
    plt.subplot(2, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
